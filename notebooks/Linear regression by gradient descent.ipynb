```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

# Linear regression by gradient descent

```python
m = 100
X = np.linspace(-5, 5, m)
```

```python
a = 3.5
b = 2.6
noise = 2*np.random.randn(100)
Y = a * X + b + noise
```

```python
plt.scatter(X, Y, color='#33669966')
```

```python
def lin(a, b, x):
    return a*x + b
```

```python
a_hat = 0
b_hat = 0
alpha = 0.01
for i in range(1000):
    a_hat = a_hat - alpha * sum((lin(a_hat, b_hat, X) - Y) * X)/m
    b_hat = b_hat - alpha * sum(lin(a_hat, b_hat, X) - Y) /m
    
    if i % 100 == 0:
        plt.scatter(X, Y, color='#33669966')
        plt.plot(X, lin(a_hat, b_hat, X), color='#33993366')
        plt.show()
```

```python
print(a_hat, b_hat)
```
