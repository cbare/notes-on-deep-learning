```python
import numpy as np
import tensorflow as tf
from tensorflow import keras
```

```python
import matplotlib.pyplot as plt
```

```python
from math import pi
```

## Plot the sine function

```python
n = 1000
```

```python
xs = np.random.uniform(0.0, 2*pi, n)
```

```python
ys = np.sin(xs)
```

```python
plt.scatter(xs,ys, color='#33669930')
```

## Single neuron model

Of course, a single neuron isn't enough, but let's try it anyway.

```python
model = tf.keras.Sequential([keras.layers.Dense(units=1, input_shape=[1])])
```

```python
model.compile(optimizer='sgd', loss='mean_squared_error')
```

```python
model.fit(xs, ys, epochs=500)
```

```python
plt.scatter(xs, ys, color='#33669930')
plt.scatter(xs, model.predict(xs), color='#339933cc')
```

A single neuron can represent a line, so we get the best line that approximates sine in this range.

## NN with 1 hidden layer

```python
model = tf.keras.Sequential([
    keras.layers.Dense(units=1, input_shape=[1]),
    keras.layers.Dense(units=12, activation='sigmoid'),
    keras.layers.Dense(units=1),
])
```

```python
model.compile(optimizer='sgd', loss='mean_squared_error')
```

```python
model.fit(xs, ys, epochs=5000)
```

```python
plt.scatter(xs, ys, color='#33669930')
plt.scatter(xs, model.predict(xs), color='#339933cc')
```

## Add some noise

Let's try adding some normally distributed noise to our target variable. It turns out, our small neural net is fairly tolerant of this kind of noise.

```python
noisy_ys = ys + np.random.standard_normal(size=n)
```

```python
plt.scatter(xs, noisy_ys, color='#33669930')
plt.scatter(xs, ys, color='#3333EE30')
```

```python
model = tf.keras.Sequential([
    keras.layers.Dense(units=1, input_shape=[1]),
    keras.layers.Dense(units=12, activation='sigmoid'),
    keras.layers.Dense(units=1),
])
```

```python
model.compile(optimizer='sgd', loss='mean_squared_error')
```

```python
model.fit(xs, noisy_ys, epochs=5000)
```

```python
plt.scatter(xs, noisy_ys, color='#33669930')
plt.scatter(xs, ys, color='#3333EE30')
plt.scatter(xs, model.predict(xs), color='#339933cc')
```

```python
plt.scatter(xs, ys, color='#3333EE30')
plt.scatter(xs, model.predict(xs), color='#339933cc')
```

## Predicting outside the training range

We've only given the model one complete cycle, so it's not really fair to expect it to learn periodicity.

```python
new_xs = np.random.uniform(-2*pi, 4*pi, 200)
```

```python
plt.scatter(xs, ys, color='#3333EE30')
plt.scatter(new_xs, model.predict(new_xs), color='#339933cc')
```
