# Neural Network Activation Functions

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

## Sigmoid

The [sigmoid](http://mathworld.wolfram.com/SigmoidFunction.html) function squashes its inputs down to between 0 and 1.
$$sigmoid(z) = \frac{1}{(1+e^{-z})}$$

```python
def sigmoid(Z):
    return 1/(1+np.exp(-Z))
```

```python
x = np.linspace(-5,5,100)
```

```python
plt.plot(x, sigmoid(x))
plt.grid(True, which='both')
```

```python
def d_sigmoid(x):
    sigma = sigmoid(x)
    return sigma * (1-sigma)
```

```python
plt.plot(x, d_sigmoid(x))
plt.grid(True, which='both')
```

## Relu

```python
def relu(Z):
    return np.maximum(0,Z)
```

```python
plt.plot(x, relu(x))
plt.grid(True, which='both')
```

The derivative of relu is just a step function, zero for negative inputs and one for positive inputs.

```python
def d_relu(x):
    return 1 * (x>0)
```

```python
plt.plot(x, d_relu(x))
plt.grid(True, which='both')
```

## Hyperbolic tangent 

Happily, the [hyperbolic tangent](http://mathworld.wolfram.com/HyperbolicTangent.html) or tanh function is built into numpy. It's similar in shape to the sigmoid, but it's output ranges between -1 and 1.

```python
plt.plot(x, np.tanh(x))
plt.grid(True, which='both')
```

It's derivative reaches a maximum of 1 at the origin.

```python
def d_tanh(x):
    return 1 - np.tanh(x)**2
```

```python
plt.plot(x, d_tanh(x))
plt.grid(True, which='both')
```
